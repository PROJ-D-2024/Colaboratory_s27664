
## 6) Prosta regresja liniowa
Chcemy przewidzieƒá **cenƒô (tys. z≈Ç)** na podstawie **metra≈ºu (m¬≤)**.
Kroki:
1. Podziel dane na cechƒô X (metra≈º) i cel y (cena).
2. Dopasuj model `LinearRegression`.
3. Obejrzyj parametry (wsp√≥≈Çczynnik i wyraz wolny).
4. Zobacz wykres dopasowania.


````python

import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# X: metra≈º (m¬≤), y: cena (tys. z≈Ç)
X = df[["metraz_m2"]].values
y = df["cena_tys"].values

model = LinearRegression()
model.fit(X, y)

print("Wsp√≥≈Çczynnik (a):", model.coef_[0])
print("Wyraz wolny (b):", model.intercept_)

# Predykcja dla kilku przyk≈Çadowych metra≈ºy
sample = np.array([[40],[60],[80],[100]])
pred = model.predict(sample)
for m2, p in zip(sample.flatten(), pred):
    print(f"Przewidywana cena dla {m2} m¬≤: {p:.1f} tys. z≈Ç")

# Wykres: punkty + linia regresji
plt.figure()
plt.scatter(X, y, label="Dane")
x_line = np.linspace(X.min(), X.max(), 100).reshape(-1,1)
y_line = model.predict(x_line)
plt.plot(x_line, y_line, label="Regresja liniowa")
plt.xlabel("Metra≈º (m¬≤)")
plt.ylabel("Cena (tys. z≈Ç)")
plt.title("Prosta regresja liniowa: cena = a*metra≈º + b")
plt.legend()
plt.show()
````


### Zadanie dla Ciebie ‚úçÔ∏è
1. Dodaj nowƒÖ kolumnƒô z **cenƒÖ za 1 m¬≤** i sprawd≈∫, jak siƒô zmienia dla r√≥≈ºnych metra≈ºy.
2. Spr√≥buj wstawiƒá do modelu drobnƒÖ nieliniowo≈õƒá (np. `metraz_m2^2` jako dodatkowƒÖ cechƒô) i zobacz, czy b≈ÇƒÖd siƒô zmniejsza.
3. Zapisz wykres do pliku PNG (np. `plt.savefig("regresja.png")`).

> Rozszerzenie: podziel dane na **train/test** (`train_test_split`) i policz metryki b≈Çƒôdu (np. MAE, RMSE).


````python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, root_mean_squared_error
import matplotlib.pyplot as plt

def train_model(X, y, val_ratio=0.2):
  val_ratio = 0.3
  X_t, X_v, y_t, y_v = train_test_split(X, y, test_size=val_ratio, random_state=42)

  model = LinearRegression()
  model.fit(X_t, y_t)

  y_pred = model.predict(X_v)
  mae = mean_absolute_error(y_v, y_pred)
  print(f"MAE: {mae}")
  rmse = root_mean_squared_error(y_v, y_pred)
  print(f"RMSE: {rmse}")

  return model, mae, rmse

````

````python

df["cena_per_m2"] = df["cena_tys"] / df["metraz_m2"]
X = df[["metraz_m2"]].values
y = df["cena_per_m2"].values

model, mae, rmse = train_model(X, y)

print("Wsp√≥≈Çczynnik (a):", model.coef_[0])
print("Wyraz wolny (b):", model.intercept_)
print(f"MAE: {mae}\tRMSE: {rmse}")

# Wykres: punkty + linia regresji
plt.figure()


plt.scatter(X, y, label="Dane")

x_plot = np.linspace(X.min(), X.max(), 100).reshape(-1,1)
y_plot = model.predict(x_plot)
plt.plot(x_plot, y_line, label="Regresja liniowa")
plt.xlabel("Metra≈º (m^2)")
plt.ylabel("Cena 1 za m^2 (tys. z≈Ç)")
plt.title("Prosta regresja liniowa: cena_per_m2 = a*metra≈º + b")
plt.legend()
plt.show()
````

````python

df["cena_per_m2"] = df["cena_tys"] / df["metraz_m2"]
df["metraz_m2_2"] = df["metraz_m2"] ** 2
X = df[["metraz_m2",
        "metraz_m2_2",
        ]].values
y = df["cena_per_m2"].values

model, mae, rmse = train_model(X, y)

print("Wsp√≥≈Çczynniki (a1, a2):", model.coef_)
print("Wyraz wolny (b):", model.intercept_)
print(f"MAE: {mae:.3f}\tRMSE: {rmse:.3f}")

plt.figure()

plt.scatter(X[:, 0], y, label="Dane", alpha=0.7)

x_line = np.linspace(X[:, 0].min(), X[:, 0].max(), 100).reshape(-1,1)
x_line_2 = x_line**2
x_line_full = np.column_stack([x_line, x_line_2])
y_line = model.predict(x_line_full)

plt.plot(x_line, y_line, color="red", label="Regresja kwadratowa")
plt.xlabel("Metra≈º (m¬≤)")
plt.ylabel("Cena za 1 m¬≤ (tys. z≈Ç)")
plt.title("Regresja kwadratowa: cena_per_m2 = a1*metra≈º + a2*metra≈º^2 + b")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
````

DodajƒÖc cechƒô nieliniowƒÖ metra≈º^2 robimy dok≈Çadnie regresje kwadratowƒÖ i po RMSE i MAE widaƒá znaczƒÖco lepsze dopasowanie modelu nieliniowego!
