
## 6) Prosta regresja liniowa
Chcemy przewidzieć **cenę (tys. zł)** na podstawie **metrażu (m²)**.
Kroki:
1. Podziel dane na cechę X (metraż) i cel y (cena).
2. Dopasuj model `LinearRegression`.
3. Obejrzyj parametry (współczynnik i wyraz wolny).
4. Zobacz wykres dopasowania.


````python

import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# X: metraż (m²), y: cena (tys. zł)
X = df[["metraz_m2"]].values
y = df["cena_tys"].values

model = LinearRegression()
model.fit(X, y)

print("Współczynnik (a):", model.coef_[0])
print("Wyraz wolny (b):", model.intercept_)

# Predykcja dla kilku przykładowych metraży
sample = np.array([[40],[60],[80],[100]])
pred = model.predict(sample)
for m2, p in zip(sample.flatten(), pred):
    print(f"Przewidywana cena dla {m2} m²: {p:.1f} tys. zł")

# Wykres: punkty + linia regresji
plt.figure()
plt.scatter(X, y, label="Dane")
x_line = np.linspace(X.min(), X.max(), 100).reshape(-1,1)
y_line = model.predict(x_line)
plt.plot(x_line, y_line, label="Regresja liniowa")
plt.xlabel("Metraż (m²)")
plt.ylabel("Cena (tys. zł)")
plt.title("Prosta regresja liniowa: cena = a*metraż + b")
plt.legend()
plt.show()
````

3. Zapisz wykres do pliku PNG (np. `plt.savefig("regresja.png")`).

> Rozszerzenie: podziel dane na **train/test** (`train_test_split`) i policz metryki błędu (np. MAE, RMSE).


````python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, root_mean_squared_error
import matplotlib.pyplot as plt

def train_model(X, y, val_ratio=0.2):
  val_ratio = 0.3
  X_t, X_v, y_t, y_v = train_test_split(X, y, test_size=val_ratio, random_state=42)

  model = LinearRegression()
  model.fit(X_t, y_t)

  y_pred = model.predict(X_v)
  mae = mean_absolute_error(y_v, y_pred)
  print(f"MAE: {mae}")
  rmse = root_mean_squared_error(y_v, y_pred)
  print(f"RMSE: {rmse}")

  return model, mae, rmse

````

````python

df["cena_per_m2"] = df["cena_tys"] / df["metraz_m2"]
X = df[["metraz_m2"]].values
y = df["cena_per_m2"].values

model, mae, rmse = train_model(X, y)

print("Współczynnik (a):", model.coef_[0])
print("Wyraz wolny (b):", model.intercept_)
print(f"MAE: {mae}\tRMSE: {rmse}")

# Wykres: punkty + linia regresji
plt.figure()


plt.scatter(X, y, label="Dane")

x_plot = np.linspace(X.min(), X.max(), 100).reshape(-1,1)
y_plot = model.predict(x_plot)
plt.plot(x_plot, y_line, label="Regresja liniowa")
plt.xlabel("Metraż (m^2)")
plt.ylabel("Cena 1 za m^2 (tys. zł)")
plt.title("Prosta regresja liniowa: cena_per_m2 = a*metraż + b")
plt.legend()
plt.show()
````

````python

df["cena_per_m2"] = df["cena_tys"] / df["metraz_m2"]
df["metraz_m2_2"] = df["metraz_m2"] ** 2
X = df[["metraz_m2",
        "metraz_m2_2",
        ]].values
y = df["cena_per_m2"].values

model, mae, rmse = train_model(X, y)

print("Współczynniki (a1, a2):", model.coef_)
print("Wyraz wolny (b):", model.intercept_)
print(f"MAE: {mae:.3f}\tRMSE: {rmse:.3f}")

plt.figure()

plt.scatter(X[:, 0], y, label="Dane", alpha=0.7)

x_line = np.linspace(X[:, 0].min(), X[:, 0].max(), 100).reshape(-1,1)
x_line_2 = x_line**2
x_line_full = np.column_stack([x_line, x_line_2])
y_line = model.predict(x_line_full)

plt.plot(x_line, y_line, color="red", label="Regresja kwadratowa")
plt.xlabel("Metraż (m²)")
plt.ylabel("Cena za 1 m² (tys. zł)")
plt.title("Regresja kwadratowa: cena_per_m2 = a1*metraż + a2*metraż^2 + b")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
````

Dodając cechę nieliniową metraż^2 robimy dokładnie regresje kwadratową i po RMSE i MAE widać znacząco lepsze dopasowanie modelu nieliniowego!
